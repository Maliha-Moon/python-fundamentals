# **Adding a new column**

## **assign()**

- **Add multiple columns in a single line of code.**
- Returns a **new DataFrame** (a copy) with the specified modifications.
- If a column **already exists**, `.assign()` **updates** its values.
- Useful for **chaining multiple operations** without modifying the original DataFrame in-place.

**Syntax:**
`new_df = df.assign(new_col1=values1, new_col2=values2, ...)`

```python
#import numpy
import numpy as np
```

```python
# create random array
code1 = np.random.randint(100,999, size=10)
code2 = np.random.randint(100,999, size=10)
```

```python
# Create series using random numbers
Series1 = pd.Series(code1, index = np.arange(0,10))
Series2 = pd.Series(code2, index = np.arange(0,10))                    
```

```python
Series2
```

```python
# using assign() to add multiple columns in a DataFrame
df = df.assign(code1 = Series1, code2 = Series2)
df
```

## **insert()**

### **Syntax**
`.insert(loc,column,value)`

### **Description:**
Inserts a new column at a specific position in a DataFrame.

### **Parameters:**
- `loc` (int) → Position index where the new column will be inserted.
- `column` (str) → Name of the new column.
- `value` (array-like) → Values for the new column (list, NumPy array, or Pandas Series).

### **Return Value:**
**None** → The method updates the existing DataFrame in-place.

```python
# using insert() add new column at a specific position
df.insert(1, 'OTP', Series2)
```
