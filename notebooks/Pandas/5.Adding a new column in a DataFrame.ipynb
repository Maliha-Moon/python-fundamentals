## **1.2 Adding new column using array**
If you want to add a new column to a DataFrame, use a **1D NumPy array** whose length matches **the number of rows** in the DataFrame

```python
# import numpy
import numpy as np
```

```python
# create a numpy array of 10 elements
customer_code = np.arange(0,10)
```

```python
# Length of the array
len(customer_code)
```

```python
# Adding a new column to DataFrame with an array
df['Customer_code'] = customer_code
```

```python
df
```
