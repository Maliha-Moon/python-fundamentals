### ðŸ“Œ Column Selection in Pandas: Series vs DataFrame

#### ðŸ”¹ `df['Name']` â†’ **Series**
- Single bracket used
- Selects one column
- Returns: `pandas.Series`

#### ðŸ”¹ `df[['Name']]` â†’ **DataFrame**
- Double brackets used
- Selects one column, but keeps it in DataFrame format
- Returns: `pandas.DataFrame`

#### ðŸ”¹ `df[['Name', 'Age']]` â†’ **DataFrame**
- Double brackets used
- Selects multiple columns
- Returns: `pandas.DataFrame`
