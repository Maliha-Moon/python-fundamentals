```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/Salary_dataset.csv')
df
```

```python
df.isnull().sum()
```

```python
df.columns
```

```python
df.drop('Unnamed: 0', axis=1)
```

```python
plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Salary dataset')
```

```python
x = df[['YearsExperience']]
x.head()
```

```python
y = df[['Salary']]
y.head()
```

```python
x.mean()
```

```python
y.mean()
```

```python
plt.scatter(x.mean(), y.mean(), color='red')
plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Salary dataset')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
```

```python
reg.fit(x,y)
```

```python
df['predicted_salary'] = reg.predict(x)
```

```python
df.head()
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y, reg.predict(x))
```

```python
plt.plot(x, reg.predict(x), color='orange')

plt.scatter(x.mean(), y.mean(), color='red')
plt.scatter(df['YearsExperience'], df['Salary'])
plt.xlabel('YearsExperience')
plt.ylabel('Salary')
plt.title('Salary dataset')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_absolute_error(df['Salary'], df['predicted_salary'])
mae
```

```python
mean_squared_error(df['Salary'], df['predicted_salary'])
```
