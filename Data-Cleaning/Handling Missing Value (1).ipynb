## **Imputing Categorical Data**

## üìù Steps

### 1Ô∏è‚É£ Identify the Data Types
- Use `.dtypes` or `info()` to check the data types of each column in the dataset.  
- This helps to distinguish between **numerical** and **categorical** features.

---

### 2Ô∏è‚É£ Handle Missing Values for Categorical Data
For categorical columns with missing values, we can apply:

- **Mode Imputation**  
  Replace missing values with the **most frequent value** (mode) of the column.

- **Forward Fill (ffill)**  
  Fill the missing value with the **previous non-missing value**.

- **Backward Fill (bfill)**  
  Fill the missing value with the **next non-missing value**.

---

### üî¢ **Mode Imputation**

**Imputation:** This technique involves filling in missing values with a **calculated estimate, such as the mean, median, or mode** of the available data. Advanced methods include **predictive modeling**, where missing values are **predicted based on relationships within the data**.

### **Mode Imputation for a Particular Column**

---

### üìù Steps:
1. **Identify** the column with missing values.  
2. **Compute the mode** (most frequent value) of that column.  
3. **Replace** all missing values (`NaN`) in the column with the computed mode. 

```python
# Select the 'Age' column & compute mode
Dataset["Age"].mode()
```

```python
# Get actual mode value
Dataset['Age'].mode()[0]
```

```python
# Fill/replace all missing values with the mode value
Dataset["Age"].fillna(Dataset["Age"].mode()[0],inplace=True)
```

### **Mode Imputation for All Object (Categorical) Columns**

---

#### üìù Steps:
1. **Identify** all object (categorical) columns in the dataset.  
2. **Detect** which of those columns contain missing values.  
3. **Compute** the mode (most frequent value) for each column with missing values.  
4. **Replace** missing values with the corresponding mode value.

```python
# collect object type data
Dataset.select_dtypes(include='object')
```

```python
# get column having missing values
Dataset.select_dtypes(include = 'object').isnull().sum()
```

```python
# list of column with object data type
Dataset.select_dtypes(include = 'object').columns
```

```python
# fill with mode value
for i in Dataset.select_dtypes(include = 'object').columns:
    Dataset[i].fillna(Dataset[i].mode()[0], inplace=True)
```

```python
Dataset.select_dtypes(include = 'object').isnull().sum() # No missing value
```
